```python
"""The shortest linear rgeression code 
line 1: imports libraries
line 2: loads data, shuffles data, and splits into train and test
line 3: computes cost using lambda function
line 4 to 9: implements gradient descent using vectorized approach, also prints cost and weights every 100 iterations
line 10: calls gradient descent function
line 11: computes predictions on cross validation set
line 12: plots predictions
line 13: defines r2 score using lambda function
line 14: prints r2 score"""
import numpy as np, pandas as pd, matplotlib.pyplot as plt
y, x_train,y_cvs,x_cvs = pd.read_csv("lineardata.csv").sample(frac=1).reset_index(drop=True).pipe(lambda df: (df.iloc[:40000,0].values, df.iloc[:40000,1:].values, df.iloc[40000:,0].values, df.iloc[40000:,1:].values))
compute_Cost = lambda x,y,w_new,b,lambda_,m: (np.sum(np.square(np.dot(x,w_new)+b-y))+np.sum(np.square(w_new))*lambda_)/(2*m)
def gradient_descent(lambda_,x,y,w,b,learning_rate,iters, m=len(y)):
    for i in range(iters):
        w-= learning_rate*((np.dot(((np.dot(x,w) +b) - y).T, x)/m).T + lambda_*np.sum(w)/m)
        b-= learning_rate*(np.sum((np.dot(x,w)+b-y))/m)
        if i%100 == 0: print(f"Cost after iteration {i}: {compute_Cost(x,y,w,b,lambda_,m)}\n w: {w}")
    return w,b
w_final,b_final = gradient_descent(0.01,x_train,y,np.zeros(20),0,0.00001,10001)
y_pred=np.dot(x_cvs,w_final)+b_final
plt.scatter(x_train[:100,0],y_pred[:100], label="Predicted"); plt.scatter(x_train[:100,0],y_cvs[:100], label="Actual"); plt.legend(); plt.show()
r2=lambda y,y_pred: 1-np.sum(np.square(y-y_pred))/np.sum(np.square(y-np.mean(y)))
print("R2 Score: ", r2(y_cvs, y_pred))
```
